```python
#Loading the libraries
import pandas as pd
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi

#Loading the data
data = pd.read_csv(r"C:\Users\Ghazy\Desktop\evansi_survival\August\span_august.csv")

#View the data
data.head(n=10)
#Summary of the variables

data.describe() #only summarize the numerical variables in the dataset
#Visualizing data

import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=data['GWSc'], y=data['Span'], showmeans=True)
plt.show()
#ANOVA F Test

model = smf.ols(formula='Span ~ GWSc', data=data)
results = model.fit()
print (results.summary())
import statsmodels.stats.multicomp as multi

test = multi.MultiComparison(data['Span'], data['GWSc'])
res = test.tukeyhsd()
print(res.summary())
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax = sns.boxplot(x='GWSc', y="Span", data=data)
ax = sns.swarmplot(x="GWSc", y="Span", data=data, color=".25", ax=ax)

ax.set(xlabel='Global warming scenario', ylabel='Lifespan [Days]')
sns.despine() #Removing axes spines
plt.ylim(0, 60)

sns.set_context("paper", font_scale=2, rc={"font.size":6,"axes.labelsize":16})
plt.show()
```

```python
#Loading the libraries
import pandas as pd
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi

#Loading the data
dat = pd.read_csv(r"C:\Users\Ghazy\Desktop\evansi_survival\lifetable_evansi_VIZ.csv")

#View the data
dat.head(n=5)
#Summary of the variables

#data.describe() #only summarize the numerical variables in the dataset
#Visualizing data
```

```python
# this is reproductive rate
Eggs = sns.factorplot(x="scen1", y="eggs", col="month1", 
                      data=dat, kind="box", size=4, aspect=1, 
                      legend=False,margin_titles=False)
(Eggs.set_axis_labels("", "Reproductive rate")
.set_xticklabels(["Current", "RCP2.6", "RCP8.6"])
.set_titles("{col_name}")
.set(ylim=(0, 200)))
#.despine(left=True)) 

plt.show()
```

```python
# This Is generation time
larvae = sns.factorplot(x="scen2", y="larvae", col="month2", 
                      data=dat, kind="box", size=4, aspect=1,
                      legend=False,margin_titles=False)
(larvae.set_axis_labels("", "Generation time")
.set_xticklabels(["Current", "RCP2.6", "RCP8.6"])
.set_titles("{col_name}")
.set(ylim=(0, 30)))
#.despine(left=True)) 

plt.show()
```

```python
# This is intrinisc rate of natural increase
proto = sns.factorplot(x="scen3", y="proto", col="month3", 
                      data=dat, kind="box", size=4, aspect=1,
                      legend=False,margin_titles=False)
(proto.set_axis_labels("", "Intrinsic rate of increase")
.set_xticklabels(["Current", "RCP2.6", "RCP8.6"])
.set_titles("{col_name}")
.set(ylim=(0, 0.5)))
#.despine(left=True)) 

plt.show()
```

```python
# this is finite rate
deuto = sns.factorplot(x="scen4", y="deuto", col="month4", 
                      data=dat, kind="box", size=4, aspect=1,
                      legend=False,margin_titles=False)
(deuto.set_axis_labels("", "Finite rate of increase")
.set_xticklabels(["Current", "RCP2.6", "RCP8.6"])
.set_titles("{col_name}")
.set(ylim=(0, 2)))
#.despine(left=True)) 

plt.show()
```

```python
# this is doubling time
total = sns.factorplot(x="scen5", y="total", col="month5", 
                      data=dat, kind="box", size=4, aspect=1,
                      legend=False,margin_titles=False)
(total.set_axis_labels("", "Doubling time")
.set_xticklabels(["Current", "RCP2.6", "RCP8.6"])
.set_titles("{col_name}")
.set(ylim=(0, 7)))
#.despine(left=True)) 

plt.show()
```
